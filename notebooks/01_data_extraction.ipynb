```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
```

```python
# list of stosks from different sectors
STOCKS = {
    "HDFC BANK": "HDFCBANK.NS",
    "TCS": "TCS.NS",
    "RELIANCE": "RELIANCE.NS",
    "HINDUSTAN UNILEVER": "HINDUNILVR.NS",
    "SUN PHARMA": "SUNPHARMA.NS"
}
```

```python
# function to load stock data
def load_stock_data(ticker, period="1y", interval="1d"):

    df = yf.download(
        ticker,
        period=period,
        interval=interval,
        auto_adjust=True,
        progress=False
    )

    #  FIX multi index columns
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df.reset_index(inplace=True)

    return df
```

```python
df = load_stock_data("RELIANCE.NS")
df.head()
```

```python
df.info()
```

```python
stock_data = []

for stock_name, ticker in STOCKS.items():
    
    temp_df = load_stock_data(ticker)
    
    temp_df["Stock"] = stock_name
    
    stock_data.append(temp_df)

master_df = pd.concat(stock_data, ignore_index=True)
```

```python
import os
# save the master_df data to data folder
os.makedirs("data", exist_ok=True)

master_df.to_csv("data/raw_stock_data.csv", index=False)
```
